```python
from collections import Counter
import heapq
def get_n_largest(input_dict, n, include_values: bool = False):
    res = []
    for key, val in sorted(input_dict.items(), key=lambda x: x[1], reverse=True)[:n]:
        if include_values:
            res.append((key, val))
        else:
            res.append(key)
    return res

diff_leagues_differentials = {}
ratio_leagues_differentials = {}

event_df = pd.read_csv(os.path.join(outputs_dir, 'open_dsml_league.csv'))

for pos_col in ('gk', 'def', 'mid', 'frw'):
    max_diff = -float('inf')
    max_ratio = -float('inf')
    
    counter = Counter(map(lambda x: x.strip(), event_df[pos_col].str.split(',').sum()))
    league_ownership_summary = {player: count / event_df.shape[0] for player, count in counter.items()}
    
    diff_dict = {}
    ratio_dict = {}
    for player, ownership in league_ownership_summary.items():
        global_ownership = float(players_info.summary.groupby('name').ownership.max()[player]) / 100 * 11 / 15
        
        diff = ownership  - global_ownership
        diff_dict[player] = diff
        
        ratio = ownership / global_ownership
        ratio_dict[player] = ratio
            
    pos2n_vis = {'gk': 2, 'def': 5, 'mid': 5, 'frw': 3}
    print(pos_col, get_n_largest(ratio_dict, pos2n_vis[pos_col]))
```

```python
from collections import Counter
import heapq
def get_n_largest(input_dict, n, include_values: bool = False):
    res = []
    for key, val in sorted(input_dict.items(), key=lambda x: x[1], reverse=True)[:n]:
        if include_values:
            res.append((key, val))
        else:
            res.append(key)
    return res

diff_leagues_differentials = {}
ratio_leagues_differentials = {}

event_df = pd.read_csv(os.path.join(outputs_dir, 'fpl_hq.csv'))
bench = Counter(map(lambda x: x.strip(), event_df['bench'].str.split(',').sum()))

for pos_col in ('gk', 'def', 'mid', 'frw'):
    max_diff = -float('inf')
    max_ratio = -float('inf')
    
    counter = Counter(map(lambda x: x.strip(), event_df[pos_col].str.split(',').sum()))
    for name, count in bench.items():
        if players_info.get_pos_by_name(name) == pos_col:
            counter[name] += count
    league_ownership_summary = {player: count / event_df.shape[0] for player, count in counter.items()}
    
    counter = Counter(league_ownership_summary).most_common(10)

    diff_dict = {}
    ratio_dict = {}
    for player, ownership in league_ownership_summary.items():
        global_ownership = float(players_info.summary.groupby('name').ownership.max()[player]) / 100 * 11 / 15
        
        diff = ownership  - global_ownership
        diff_dict[player] = diff
        
        ratio = ownership / global_ownership
        ratio_dict[player] = ratio
            
    pos2n_vis = {'gk': 2, 'def': 5, 'mid': 5, 'frw': 3}
    print(pos_col, get_n_largest(ratio_dict, pos2n_vis[pos_col]))
```
