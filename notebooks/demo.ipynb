```python
from IPython.display import display, HTML


class FantasyInfo():
    def __init__(self, leagues_list):
        self.leagues_list = leagues_list
        self.users_cache = {}
        self.leagues_cache = {}
    
    @classmethod
    def _get_init_standings(cls, league_id, page: int = 1):
        standings_body = "https://fantasy.premierleague.com/api/leagues-classic"
        standings_suffix = f"standings/?page_standings={page}"
        url = os.path.join(standings_body, str(league_id), standings_suffix)
        data = safe_request(url)
        league_name = data['league']['name']
        standings = data['standings']['results']
        if len(standings) == 50:
            print('Make additional league query')
            _, tail = cls._get_init_standings(league_id, page + 1)
            standings.extend(tail)
        return league_name, standings
    
    @staticmethod
    def _get_user_info(entry):
        personal_body = "https://fantasy.premierleague.com/api/entry/"
        personal_suffix = "history"
        url = os.path.join(personal_body, str(entry), personal_suffix)
        data = safe_request(url)
        return data['current']
    
    def _beautify_standings(self, standings):
        cols_order = ['rank', 'entry_name', 'player_name', 'total', 'last_rank', 'entry']
        return pd.DataFrame(standings)[cols_order].set_index('rank')
    
    def _cache_entries(self, entries):
        for entry in tqdm(entries):
            if entry in self.users_cache:
                continue
            else:
                user_info = self._get_user_info(entry)
                self.users_cache[entry] = user_info
            
    def _get_gw_history(self, entries):
        records = []
        for entry in entries:
            gws = self.users_cache[entry]
            record = {'entry': entry}
            for gw in gws:
                event_id = gw['event']
                record[f'gw_{event_id}_points'] = gw['points'] - gw['event_transfers_cost']
                record[f'gw_{event_id}_hits'] = gw['event_transfers_cost']
#             record['total_points'] = gw['total_points']
            record[f'bank'] = gw['bank'] / 10
            record[f'global_rank'] = int(gw['overall_rank'])    
            records.append(record)
        return pd.DataFrame(records).fillna(0)
    
    def _mean_of_top25(scores):
        top_25 = len(scores) // 4
        top_scores = sorted(scores)[-top_25:]
        return np.mean(top_scores)

    def update(self):
        for league_id in self.leagues_list:
            league_name, standings = self._get_init_standings(league_id)
            print(league_name)
            standings_table = self._beautify_standings(standings)
            entries = standings_table.entry.tolist()
            self._cache_entries(entries)
            history = self._get_gw_history(entries)
            self.leagues_cache[league_name] = pd.merge(standings_table, history, on='entry')
            display(self.leagues_cache[league_name])

#     def leagues_leaderboard(self):
        
leagues_list = [
#     579956,
    356774,
#     406089, 626932, 5712, 458888, 406273, 257622, 180769, 684405, 381725, 82273
]
fantasy_info = FantasyInfo(leagues_list)
fantasy_info.update()
```

```python
def mean_of_top_percentille(scores, percentille):
    percentille_value = np.percentile(scores, percentille)
    top_scores = scores[scores > percentille_value]
    return np.mean(top_scores)

records = []
for league_name, league_standings in fantasy_info.leagues_cache.items():
    print(league_name)
    record = {'league_name': league_name, 'n_players': league_standings.shape[0]}
    for gw in range(1,5):
        gw_scores = league_standings[f'gw_{gw}_points']
        gw_scores = gw_scores[gw_scores > 0].values
        record[f'mean_gw_{gw}'] = np.mean(gw_scores)
#         record[f'median_gw_{gw}'] = np.median(gw_scores)
#         record[f'mean_top_25_gw_{gw}'] = mean_of_top_percentille(gw_scores, 25)
#         record[f'mean_top_50_gw_{gw}'] = mean_of_top_percentille(gw_scores, 50)
    records.append(record)
leaderboard = pd.DataFrame(records).set_index('league_name')

base_cols = leaderboard.columns
leaderboard['total'] = leaderboard[base_cols].sum(axis=1)

for col in base_cols:
    leaderboard[f'{col}_rank'] = leaderboard[col].rank()

rank_cols = [col for col in leaderboard.columns if 'rank' in col]

leaderboard['total_rank'] = leaderboard[rank_cols].sum(axis=1)
leaderboard.sort_values('total_rank', ascending=False)
```
